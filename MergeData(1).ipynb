```python
import pandas as pd
import numpy as np
info = pd.read_csv("Final Cleaning Data(1).csv")
review = pd.read_csv("newreview.csv")

info = info.rename(columns={"drama_name": "title"})
info.head(3)
```

```python
print(review.head(3))
result = pd.merge(info, review, on="title", how="outer")
```

```python
result.to_csv('MergedData.csv')
```

```python

resultNew = result.groupby(['title', 'year','kdrama_id','director','screenwriter','tot_eps','duration','start_dt','end_dt','aired_on','org_net','content_rt','synopsis','rank','pop'])[['music_score', 'story_score','acting_cast_score','rewatch_value_score','overall_score','n_helpful']].mean()
```

```python
resultNew.head(3)
```

```python
columns_to_round = ['music_score', 'story_score','acting_cast_score','rewatch_value_score','overall_score']

# Round the specified columns to 2 decimal places
resultNew[columns_to_round] = resultNew[columns_to_round].round(1)

resultNew['n_helpful'] = resultNew['n_helpful'].round(0)
```

```python
resultNew.head(3)
```

```python
df_cleaned = resultNew.dropna()
```

```python
df_sorted_desc = df_cleaned.sort_values(by='year', ascending=True)
```

```python
df_sorted_desc.to_csv('Final CLeaned Dataset.csv')
```
